```python
import numpy as np
import pandas as pd
```

Read in your label data:

```python
labels = pd.read_csv('labels.csv')
labels.head(10)
```

```python
# Let's conver benign to 0 and cancer to 1 on the whole dataset
labels = labels.replace({'benign': 0, 'cancer': 1})
labels.head(10)
```

## Create your first ground truth as derived from biopsy labels: 

```python
gt1 = labels['biopsy']
gt1.head(10)
```

## Create your second truth by voting system from the three radiologists:

```python
# Define gt2 as the voting system from the three radiologists
gt2 = labels[['rad1','rad2','rad3']].sum(axis=1)
gt2 = (gt2 > 1).replace(True,1).replace(False,0)
gt2.head(10)
```

## Create your third ground truth by weighting the three radiologists:

```python
# rad1 has 5 years experience, rad2 has 10 years experience, and rad3 has 15 years experience
rad1_weight = 5/30
rad2_weight = 10/30
rad3_weight = 15/30

labels_weighted = labels.copy()

labels_weighted['rad1'] = labels_weighted['rad1']*rad1_weight
labels_weighted['rad2'] = labels_weighted['rad2']*rad2_weight
labels_weighted['rad3'] = labels_weighted['rad3']*rad3_weight

gt3 = labels_weighted[['rad1','rad2','rad3']].sum(axis=1)
gt3 = (gt3 > 1).replace(True,1).replace(False,0)
gt3.head(10)
```

## Compare the three ground truths:

Here, just explore the three sets of labels you created and see how often they agree

```python
# Accuracy of gt1
gt1_accuracy = (gt2 == gt1).mean()
print("Accuracy of gt1 (simple voting):")
print(gt1_accuracy)

# Accuracy of gt2
gt2_accuracy = (gt3 == gt1).mean()
print("Accuracy of gt2 (weighted voting):")
print(gt2_accuracy)
```
